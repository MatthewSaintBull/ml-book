```python
def stimami_la_RAL(linguaggio, is_manager, istruzione, anni_esp):
    ral = minimo
    
    # aggiungo il peso riferito al linguaggio
    ral += ral * pesi_linguaggio[linguaggio]
    
    # aggiungo il peso riferito agli anni di esperienza
    ral += ral * pesi_anni[anni_esp]
    
    # aggiungo il peso riferito al grado di istruzione
    ral += ral * pesi_istruzione[istruzione] 
    
    # Ã¨ un responsabile?
    if is_manager:
        ral += ral * peso_manager
        
    return ral
    
```

```python
pesi_linguaggio = {
    'Python': 0.05,
    'Javascript': 0.01,
    'Java': 0.03,
    'R': 0.07,
    'C#': 0.02,
    'Objective-C/Swift': 0.03,
    'SQL': 0.02,
    'COBOL': 0.04
}

pesi_istruzione = {
    'Diploma': 0,
    'Laurea Triennale': 0.02,
    'Laurea Magistrale': 0.03,
    'Dottorato': 0.05
}

def pesi_anni(anni_di_esperienza):
    if anni_di_esperienza > 8:
        return 1
    elif 6 < anni_di_esperienza <= 8:
        return 0.7
    elif 4 < anni_di_esperienza <= 6:
        return 0.5
    elif 1 < anni_di_esperienza <= 4:
        return 0.25
    else:
        return 0
    
    
pesi_anni = {anni: pesi_anni(anni) for anni in range(1, 41)}

peso_manager = 0.5
```

```python
def calcola_stima(riga):
    ral = stimami_la_RAL(
        riga['Linguaggio'], 
        riga['Responsabile'], 
        riga['Istruzione'], 
        riga['Esperienza']
    )
    return ral


df['Stima'] = df.apply(calcola_stima, axis=1)
df.head()
```

```python
def costo(df):
    errori = []
    for indice, riga in df.iterrows():
        errori.append((riga['Stima'] - riga['RAL'])**2)
    costo = sum(errori) / (2*len(df))
    return costo

costo(df)
```

```python
from sklearn import preprocessing

linguaggi_enc = preprocessing.LabelEncoder()
istruzione_enc = preprocessing.LabelEncoder()

linguaggi = linguaggi_enc.fit(df['Linguaggio'])
istruzione = istruzione_enc.fit(df['Istruzione'])    

df['Linguaggio'] = linguaggi.transform(df['Linguaggio'])
df['Istruzione'] = istruzione.transform(df['Istruzione'])
```

```python
def gradientDescent(X, y, theta, alpha, m, numIterazioni):
    # preparo la trasposta di X
    xTrans = X.transpose()
    
    ciclo_costi = []
    
    for i in range(0, numIterazioni):
        ral = 20000  # minimo
        
        # calcolo la stima con i pesi attuali
        stima = ral + np.dot(X, theta)
        
        # l'errore
        errore = stima - y
        
        # e il costo totale
        costo = np.sum(errore ** 2) / (2 * m)
        
        ciclo_costi.append((i, costo))
        
        # calcolo il gradiente e aggiorno i pesi
        gradiente = np.dot(xTrans, errore) / m
        theta = theta - alpha * gradiente
        
    return ciclo_costi, theta
```

```python
righe = df[[
    'Linguaggio', 
    'Esperienza', 
    'Istruzione', 
    'Responsabile']].as_matrix()
y = df['RAL'].as_matrix()
m, n = np.shape(righe)
theta = np.ones(n)
alpha = 0.05
max_iter = 1000
```

```python
costi, pesi = gradientDescent(righe, y, theta, alpha, m, max_iter)
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig, ax1 = plt.subplots(figsize=(15,10), 
                        subplot_kw={'projection': '3d'})

alpha = 0.8
r = np.linspace(-alpha,alpha,100)
X,Y= np.meshgrid(r,r)

def func(X, Y):
    return X**2+Y**2

l = func(X, Y)

ax1.plot_surface(X,Y,l)
ax1.set_xlabel('\n$\\theta_{0}$', fontsize=20)
ax1.set_ylabel('\n$\\theta_{1}$', fontsize=20)
ax1.set_zlabel('\n$J(\\theta_{0},\\theta_{1})$', fontsize=20)

fig.savefig('cost.png', bbox_inches='tight')
plt.show()
```
