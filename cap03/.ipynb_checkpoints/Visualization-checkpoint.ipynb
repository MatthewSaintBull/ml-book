```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

def funzione_retta(x):
    return 3*x+1

X = np.arange(10)
y = funzione_retta(X)

fig = plt.figure()
plt.plot(X, y, label='y=3x+1')
plt.plot(X, np.sin(X), label='y=sin(x)')
plt.legend()
fig.savefig('retta5.png')
```

```python
import pandas as pd

df = pd.DataFrame(
    {
        'tipo': [
            'Sintassi',
            'PEP8',
            'Bug',
            'Performance',
            'Funzioni Lunghe',
            'Codice non documentato',
            'Naming'
        ],
        'count': [0, 2, 2, 1, 2, 3, 3]
    },
    columns = ['tipo', 'count']
)
print(df)
```

```python
fig = plt.figure()
plt.bar(range(len(df)), df['count'], align='center')
fig.savefig('bar1.png')
```

```python
fig = plt.figure()
plt.bar(range(len(df)), df['count'], align='center')
plt.xticks(range(len(df)), df['tipo'])
fig.savefig('bar2.png')
```

```python
fig = plt.figure()
plt.bar(range(len(df)), df['count'], align='center')
plt.xticks(range(len(df)), df['tipo'], rotation='vertical')
fig.savefig('bar3.png', bbox_inches='tight')
```

```python
fig = plt.figure()
plt.bar(range(len(df)), df['count'], align='center', color='red')
plt.xticks(range(len(df)), df['tipo'], rotation='vertical')
fig.savefig('bar4.png', bbox_inches='tight')
```

```python
fig = plt.figure()
colors = ['white', 'b', 'red', '#624ea7', 'k', 'cyan', '#009966']
plt.bar(range(len(df)), df['count'], align='center', color=colors)
plt.xticks(range(len(df)), df['tipo'], rotation='vertical')
fig.savefig('bar5.png', bbox_inches='tight')
```
